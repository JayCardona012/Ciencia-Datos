```python
customers = shopping[['customer_id', 'gender', 'age']].drop_duplicates()
customers.rename(columns={'customer_id': 'customers_id'}, inplace=True)
customers
```

```python
categories = shopping[['category']].drop_duplicates()
categories['category_id'] = range(1, len(categories) + 1)
categories.rename(columns={'category': 'category_name'}, inplace=True)
categories
```

```python
payments = shopping[['payment_method']].drop_duplicates()
payments['payment_id'] = range(1, len(payments) + 1)
payments.rename(columns={'payment_method': 'payment_method'}, inplace=True)
payments
```

```python
malls = shopping[['shopping_mall']].drop_duplicates()
malls['shopping_mall_id'] = range(1, len(malls) + 1)
malls.rename(columns={'shopping_mall': 'mall_name'}, inplace=True)
malls
```

```python
shopping['invoice_date'] = pd.to_datetime(shopping['invoice_date'], dayfirst=True)
dates = shopping[['invoice_date']].drop_duplicates()
dates['invoice_date_id'] = range(1, len(dates) + 1)
dates['year'] = dates['invoice_date'].dt.year
dates['month'] = dates['invoice_date'].dt.month
dates['day'] = dates['invoice_date'].dt.day
dates.rename(columns={'invoice_date': 'full_date'}, inplace=True)
dates
```

```python
invoices = shopping.merge(customers, left_on='customer_id', right_on='customers_id') \
                  .merge(categories, left_on='category', right_on='category_name') \
                  .merge(payments, left_on='payment_method', right_on='payment_method') \
                  .merge(malls, left_on='shopping_mall', right_on='mall_name') \
                  .merge(dates, left_on='invoice_date', right_on='full_date')
                  
invoices = invoices[['invoice_no', 'customers_id', 'category_id', 'payment_id', 'invoice_date_id', 'shopping_mall_id', 'quantity', 'price']]
invoices['total_price'] = invoices['quantity'] * invoices['price']
invoices.rename(columns={'invoice_no': 'invoice_detail_id'}, inplace=True)

invoices
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
ventas_por_categoria = shopping.groupby("category")["price"].sum().reset_index()
ventas_por_categoria.rename(columns={"price": "total_sales"}, inplace=True)


plt.figure(figsize=(10, 6))
sns.barplot(x="category", y="total_sales", data=ventas_por_categoria, palette="viridis")
plt.xlabel("Total Ventas")
plt.ylabel("CategorÃ­a")
plt.title("Total de Ventas por CategorÃ­a de Producto")
plt.show()
```

```python
clientes_top = shopping.groupby("customer_id")["price"].sum().reset_index()
clientes_top.rename(columns={"price": "total_spent"}, inplace=True)
clientes_top = clientes_top.sort_values("total_spent", ascending=False).head(10)

# ðŸ“Š GrÃ¡fico de barras horizontal - Clientes con mÃ¡s compras
plt.figure(figsize=(10, 5))
sns.barplot(x="total_spent", y="customer_id", data=clientes_top, palette="coolwarm", legend=True)
plt.xlabel("Total Gastado")
plt.ylabel("Cliente ID")
plt.title("Top 10 Clientes con Mayor Volumen de Compras")
plt.show()
```

```python
metodos_pago = shopping["payment_method"].value_counts().reset_index()
metodos_pago.columns = ["payment_method", "usage_count"]

# ðŸ¥§ GrÃ¡fico de torta - MÃ©todos de pago mÃ¡s usados
plt.figure(figsize=(7, 7))
plt.pie(metodos_pago["usage_count"], labels=metodos_pago["payment_method"], autopct="%1.1f%%", colors=sns.color_palette("pastel"))
plt.title("DistribuciÃ³n de MÃ©todos de Pago")
plt.show()
```

```python
# Convertir invoice_date a tipo datetime
shopping["invoice_date"] = pd.to_datetime(shopping["invoice_date"], format="%d/%m/%Y")

ventas_por_mes = shopping.groupby(shopping["invoice_date"].dt.to_period("M"))["price"].sum().reset_index()
ventas_por_mes.rename(columns={"price": "total_sales"}, inplace=True)
ventas_por_mes["invoice_date"] = ventas_por_mes["invoice_date"].astype(str)  # Convertir a string para graficar

# ðŸ“ˆ GrÃ¡fico de lÃ­neas - Ventas por mes
plt.figure(figsize=(10, 5))
sns.lineplot(x="invoice_date", y="total_sales", data=ventas_por_mes, marker="o", linestyle="-")
plt.xticks(rotation=45)
plt.xlabel("Mes")
plt.ylabel("Total Ventas")
plt.title("Tendencia de Ventas por Mes")
plt.show()
```
